
## 2. Importing and creating an object

```python
import requests
from io import BytesIO
from PIL import Image
from matplotlib import pyplot as plt
from src import ProductSearchEngine
import warnings

warnings.filterwarnings("ignore")
```

```python
search_engine = ProductSearchEngine()
```

```python
def plot(img_arr):
  plt.figure(figsize=(10, 5))
  for i in range(1,11):
      plt.subplot(2, 5, i)
      plt.imshow(img_arr[i-1])
      plt.axis("off")

  plt.tight_layout()
  plt.show()
```

### 2.1 Text search

```python
output = search_engine.text_search("white men's shirt")
plot(output)
```

```python
output = search_engine.text_search("men's leather belt")
plot(output)
```

```python
output = search_engine.text_search("red saree")
plot(output)
```

### 2.2 Image search:
The printed scores for each input are the classification scores obtained from ViLT model.

```python
url = "https://m.media-amazon.com/images/I/71qpXciYP1L.jpg"
response = requests.get(url)
img = Image.open(BytesIO(response.content)).resize((224, 224))
img
```
